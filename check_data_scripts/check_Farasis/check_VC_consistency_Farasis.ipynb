```python
from pathlib import Path
from typing import Optional
import pickle

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


def _find_repo_root(start: Path) -> Path:
    start = start.resolve()
    for p in [start, *start.parents]:
        if (p / "dataset").exists() and (p / "check_data_scripts").exists():
            return p
    raise FileNotFoundError("Cannot locate project root containing dataset/ and check_data_scripts/")


def _build_time_in_seconds(df: pd.DataFrame) -> np.ndarray:
    if "time_stamp" in df.columns:
        ts = pd.to_datetime(df["time_stamp"], errors="coerce")
        if ts.notna().sum() >= 2:
            seconds = (ts - ts.iloc[0]).dt.total_seconds().to_numpy(dtype=float)
            return np.nan_to_num(seconds, nan=0.0)

    if "del_time" in df.columns:
        dt = pd.to_numeric(df["del_time"], errors="coerce").fillna(0.0).to_numpy(dtype=float)
        dt = np.maximum(dt, 0.0)
        return np.cumsum(dt)

    return np.arange(len(df), dtype=float)


def _normalize_time_axis(t: np.ndarray) -> np.ndarray:
    t = np.asarray(t, dtype=float)
    if len(t) == 0:
        return t

    t = np.nan_to_num(t, nan=0.0)
    t = t - float(t[0])
    return np.maximum.accumulate(t)


def extract_before_from_parquet(parquet_path: Path):
    cols = ["current (A)", "voltage (V)", "time_stamp", "del_time"]
    df = pd.read_parquet(parquet_path, columns=cols).copy()

    current = pd.to_numeric(df["current (A)"], errors="coerce").to_numpy(dtype=float)
    voltage = pd.to_numeric(df["voltage (V)"], errors="coerce").to_numpy(dtype=float)
    time_abs = _build_time_in_seconds(df)

    valid = np.isfinite(current) & np.isfinite(voltage) & np.isfinite(time_abs)
    current = current[valid]
    voltage = voltage[valid]
    time_abs = _normalize_time_axis(time_abs[valid])

    if len(time_abs) < 2:
        return np.array([]), np.array([]), np.array([])

    return time_abs, current, voltage


def extract_after_from_pkl(pkl_path: Path, first_n_cycles: Optional[int] = None):
    with pkl_path.open("rb") as f:
        obj = pickle.load(f)

    cycle_data = obj.get("cycle_data", [])
    if first_n_cycles is not None:
        cycle_data = cycle_data[:first_n_cycles]

    t_list, i_list, v_list = [], [], []
    offset = 0.0

    for cyc in cycle_data:
        t = np.asarray(cyc.get("time_in_s", []), dtype=float)
        i = np.asarray(cyc.get("current_in_A", []), dtype=float)
        v = np.asarray(cyc.get("voltage_in_V", []), dtype=float)

        valid = np.isfinite(t) & np.isfinite(i) & np.isfinite(v)
        t, i, v = t[valid], i[valid], v[valid]
        if len(t) < 2:
            continue

        t = _normalize_time_axis(t) + offset
        offset = float(t[-1])

        t_list.append(t)
        i_list.append(i)
        v_list.append(v)

    if len(t_list) == 0:
        return np.array([]), np.array([]), np.array([])

    t = _normalize_time_axis(np.concatenate(t_list))
    i = np.concatenate(i_list)
    v = np.concatenate(v_list)
    return t, i, v


def _plot_before_after(bt, bi, bv, at, ai, av, dv, di, stem: str, max_plot_points: Optional[int] = None):
    n = len(bt)
    if n == 0:
        return

    # Plot full data by default; only downsample if an explicit limit is provided.
    if max_plot_points is None or max_plot_points <= 0 or n <= max_plot_points:
        idx = np.arange(0, n, dtype=int)
    else:
        step = max(1, int(np.ceil(n / max_plot_points)))
        idx = np.arange(0, n, step, dtype=int)

    bt_p, bi_p, bv_p = bt[idx], bi[idx], bv[idx]
    at_p, ai_p, av_p = at[idx], ai[idx], av[idx]
    dv_p, di_p = dv[idx], di[idx]

    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 9), sharex=False)
    ax1.plot(bt_p, bv_p, label="Before (parquet)", linewidth=1.2)
    ax1.plot(at_p, av_p, label="After (pkl)", linewidth=1.0, alpha=0.8)
    ax1.set_title(f"Voltage vs Time (Before vs After) | {stem}")
    ax1.set_xlabel("Time(s)")
    ax1.set_ylabel("Voltage(V)")
    ax1.legend()
    ax1.grid(alpha=0.3)

    ax2.plot(bt_p, bi_p, label="Before (parquet)", linewidth=1.2)
    ax2.plot(at_p, ai_p, label="After (pkl)", linewidth=1.0, alpha=0.8)
    ax2.set_title(f"Current vs Time (Before vs After) | {stem}")
    ax2.set_xlabel("Time(s)")
    ax2.set_ylabel("Current(A)")
    ax2.legend()
    ax2.grid(alpha=0.3)

    fig.tight_layout()
    plt.show()
    plt.close(fig)

    fig2, (bx1, bx2) = plt.subplots(2, 1, figsize=(14, 8), sharex=True)
    bx1.plot(idx, dv_p, linewidth=0.8)
    bx1.set_ylabel("dV")
    bx1.grid(alpha=0.3)

    bx2.plot(idx, di_p, linewidth=0.8)
    bx2.set_ylabel("dI")
    bx2.set_xlabel("Point Index")
    bx2.grid(alpha=0.3)

    fig2.tight_layout()
    plt.show()
    plt.close(fig2)


def compare_before_after(
    parquet_path: Path,
    pkl_path: Path,
    make_plots: bool = False,
    max_plot_points: Optional[int] = None,
    first_n_cycles: Optional[int] = None,
):
    bt, bi, bv = extract_before_from_parquet(parquet_path)
    at, ai, av = extract_after_from_pkl(pkl_path, first_n_cycles=first_n_cycles)

    n = min(len(bt), len(at))
    if n == 0:
        raise ValueError("No valid points found for comparison.")

    bt, bi, bv = bt[:n], bi[:n], bv[:n]
    at, ai, av = at[:n], ai[:n], av[:n]

    dv = bv - av
    di = bi - ai

    report = {
        "points_before": int(len(bt)),
        "points_after": int(len(at)),
        "same_length": bool(len(bt) == len(at)),
        "voltage_exact_equal": bool(np.array_equal(bv, av)),
        "current_exact_equal": bool(np.array_equal(bi, ai)),
        "voltage_allclose_1e-6": bool(np.allclose(bv, av, atol=1e-6, rtol=1e-6)),
        "current_allclose_1e-6": bool(np.allclose(bi, ai, atol=1e-6, rtol=1e-6)),
        "max_abs_err_voltage": float(np.max(np.abs(dv))),
        "max_abs_err_current": float(np.max(np.abs(di))),
    }

    if make_plots:
        _plot_before_after(bt, bi, bv, at, ai, av, dv, di, pkl_path.stem, max_plot_points=max_plot_points)

    return report


# ===============================
# Batch run config (speed-first)
# ===============================
FIRST_N_CYCLES = None         # None 表示使用全部 cycle_data
MAX_FILES = None              # None 表示处理全部
PLOT_FIRST_N = 10             # 仅前 N 个样本画图，0 表示不画
PLOT_EVERY_N = 0              # 每 N 个样本画一次图，0 表示不启用
MAX_PLOT_POINTS = None        # None 表示画全量点，不做下采样
LOG_EVERY_N = 10              # 每 N 个样本输出一次进度

repo_root = _find_repo_root(Path.cwd())
raw_root = repo_root / "dataset" / "raw" / "Farasis"
pkl_dir = repo_root / "dataset" / "processed" / "Farasis"

# 构建 parquet 映射：Farasis_<raw相对路径去后缀并下划线拼接> -> parquet 文件
parquet_map = {}
for pq in raw_root.rglob("*.parquet"):
    rel_no_suffix = pq.relative_to(raw_root).with_suffix("")
    key = f"Farasis_{'_'.join(rel_no_suffix.parts)}"
    parquet_map[key] = pq

# 遍历全部 Farasis pkl（按文件名顺序）
pkl_files = sorted([p for p in pkl_dir.glob("Farasis_*.pkl")])
if MAX_FILES is not None:
    pkl_files = pkl_files[:MAX_FILES]
if len(pkl_files) == 0:
    raise FileNotFoundError(f"No Farasis pkl files found under: {pkl_dir}")

ok = 0
missing_parquet = 0
failed = 0

for i, pkl_path in enumerate(pkl_files, start=1):
    stem = pkl_path.stem
    parquet_path = parquet_map.get(stem)
    if parquet_path is None:
        missing_parquet += 1
        print(f"[{i}/{len(pkl_files)}] SKIP missing parquet: {stem}")
        continue

    make_plots = (PLOT_FIRST_N > 0 and i <= PLOT_FIRST_N) or (PLOT_EVERY_N > 0 and i % PLOT_EVERY_N == 0)

    try:
        report = compare_before_after(
            parquet_path,
            pkl_path,
            make_plots=make_plots,
            max_plot_points=MAX_PLOT_POINTS,
            first_n_cycles=FIRST_N_CYCLES,
        )
        ok += 1
        print(f"[{i}/{len(pkl_files)}] OK   {stem} | max_dV={report['max_abs_err_voltage']:.6g}, max_dI={report['max_abs_err_current']:.6g}")
    except Exception as e:
        failed += 1
        print(f"[{i}/{len(pkl_files)}] FAIL {stem} -> {type(e).__name__}: {e}")

    if i % LOG_EVERY_N == 0 or i == len(pkl_files):
        print(f"Progress: {i}/{len(pkl_files)}")

print("\nDone")
print(f"total_selected={len(pkl_files)}")
print(f"ok={ok}")
print(f"missing_parquet={missing_parquet}")
print(f"failed={failed}")
```
